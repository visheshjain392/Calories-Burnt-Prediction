Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Data Collection & Processing

```python
# loading the data from csv file to a Pandas DataFrame
calories = pd.read_csv('/content/calories.csv')
```

```python
# print the first 5 rows of the dataframe
calories.head()
```

```python
exercise_data = pd.read_csv('/content/exercise.csv')
```

```python
exercise_data.head()
```

Combining the two DataFrames

```python
calories_data = pd.concat([exercise_data, calories['Calories']], axis=1)
```

```python
calories_data.head()
```

```python
# checking the number of rows and columns
calories_data.shape
```

```python
# getting some informations about  the data
calories_data.info()
```

```python
# checkinng for missing values
calories_data.isnull().sum()
```

Data Analysis

```python
# get some statistical measures about the data
calories_data.describe()
```

Data Visualization

```python
sns.set()
```

```python
# plotting the gender columns in count plot
sns.countplot(calories_data['Gender'])
```

```python
# finding the distribution of 'Age' column
sns.distplot(calories_data['Age'])
plt.show()
```

```python
# finding the distribution of 'Height' column
sns.distplot(calories_data['Height'])
plt.show()
```

```python
# finding the distribution of 'weight' column
sns.distplot(calories_data['Weight'])
plt.show()
```

Finding the Corretation in the dataset

1. Positive Correlation
2. Negetive Correlation

```python
correlation = calories_data.select_dtypes(include=['number']).corr()
print(correlation)
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Converting the text data to numerical values

```python
calories_data.replace({'Gender':{'male':0,'female':1}}, inplace=True)
```

```python
calories_data.head()
```

Separating features and Target

```python
X = calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into trainint data and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# trainning the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on Test Data

```python
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

Mean Absolute Error

```python
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)
print("The Mean Absolute Error = ", mae)
```
